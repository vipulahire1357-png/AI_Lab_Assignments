# **# Rule Based Expert System for Career Recommendation**

An expert system is a computer program that mimics human experts. It asks questions, takes your answers, and gives advice (just like a human career counselor or doctor). Expert systems usually:

- Take inputs (user answers).

- Apply rules (if-then logic).

- Give an output (suggestion or solution).

This code is a simple expert system to recommend a career based on your interests.



```python
def get_user_input(): # This function asks quetions and collects answer.
    print("Answer with 'yes' or 'no':")
    interests = {}  # Interests = {} (Creates empty dictionary)
    interests["math"] = input("Do you like math? ").lower() == 'yes'
    interests["problem_solving"] = input("Do you enjoy problem-solving? ").lower() == 'yes'
    interests["art"] = input("Are you interested in art? ").lower() == 'yes'
    interests["creativity"] = input("Do you consider yourself creative? ").lower() == 'yes'
    interests["helping_others"] = input("Do you enjoy helping others? ").lower() == 'yes'
    interests["biology"] = input("Are you interested in biology? ").lower() == 'yes'
    interests["writing"] = input("Do you like writing? ").lower() == 'yes'
    interests["communication"] = input("Are you good at communication? ").lower() == 'yes'
    interests["logic"] = input("Do you enjoy logical thinking? ").lower() == 'yes'
    interests["coding"] = input("Do you enjoy coding? ").lower() == 'yes'
    return interests

def recommend_career(interests):  # This function applies rules to recommned answer.
    # Rule 1
    if interests["math"] and interests["problem_solving"]:
        return "Recommended Career: Engineer"
    # Rule 2
    elif interests["art"] and interests["creativity"]:
        return "Recommended Career: Designer"
    # Rule 3
    elif interests["helping_others"] and interests["biology"]:
        return "Recommended Career: Doctor"
    # Rule 4
    elif interests["writing"] and interests["communication"]:
        return "Recommended Career: Journalist"
    # Rule 5
    elif interests["logic"] and interests["coding"]:
        return "Recommended Career: Software Developer"
    else:
        return "Unable to recommend a career based on given answers. Try exploring more fields."

def main(): # This is the main function that runs the entire program.
    print("\n=== Career Recommendation Expert System ===")
    user_interests = get_user_input()
    result = recommend_career(user_interests)
    print("\n" + result)

if __name__ == "__main__":  # Run the main() function if this file is started directly by the user.
    main()
```
